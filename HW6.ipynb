## 0. Explain the difference between hypothesis testing and confidence intervals?

Before we note the difference between hypothesis testing and confidence intervals we have to mention the similarities between both which is that both are inferential methods that approximate sampling distributions using the data from a sample. The mean difference between the two methods is that hypothesis testing main goals is to relate a single conclusion of statistical significance against no statistical significance while confidence intervals main goal is to provide a range of probable values for your population. Hypothesis testng is commonly used when we want to do a precise comparison with predefined hypothesized parameter and significance level while confidence interval are used to describe a single sample.
