The first step is to set the hypothesis for the f-test since we are testing if the difference between the two variances come from the same population

$

\\

$

Now we use R to find the critical value for the f test

\
In other words, the population variances for the two sample populations come from the same population
\
Now that we know that the samples come from the same population we can use t-statistics
\
Before we do that, we have to do set the hypothesis for this test
\
$
H_0: \mu_1 = \mu_2
\\
H_a: \mu_1 < \mu_2
$
\
For this question we have to find pooled standard deviation since we the two sample come from the same population
\
$
S_p=\sqrt{\frac{(n_1-1)s_1 ^2 + (n_2-1)s_2 ^2}{n_1 + n_2 - 2}}
\\
S_p={\frac{(10-1)*78.90 ^2 + (10-1)* 96.01 ^2}{10 + 10 - 2}}
\\
S_p=87.455
$
\
Now we use the t-test for two samples
\
$
t=\frac{(\bar{x_1} - \bar{x_2})-(\mu_1 - \mu_2)}{s_p * \sqrt{\frac{1}{n_1}+\frac{1}{n_2}}}
\\
t=\frac{(115.7 - 119.7)-(0)}{87.455 * \sqrt{\frac{1}{10}+\frac{1}{10}}}
\\
t=-0.10
$
\
Now we use R to find the p-value given the t statistics
\
The p-value is 0.46072 which is way above the alpha of 0.05
\
Therefore we fail to reject the null hypothesis and there is no difference between the two means

## **4. A reproductive physiologist studying the effect of photoperiod on reproductive readiness in Japanese quail measured the cloacal gland widths of male quail exposed for 2 weeks to different day lengths:**

> |   | Long days | Short days 
> | :---: | :---: | :---:

### **Long days consisted of 16 hours of light and 8 hours of dark. Short days had 8 hours of daylight and 16 hours of dark. Assuming cloacal gland widths are normally distributed, are the glands significantly wider during long days?**

Step 1
\
Determine if this two samples come from the same population using an F-test

$
H_0: \sigma_1 = \sigma_2
\\
H_a:  \sigma_1 \ne \sigma_2

\
$

F=\frac{2.5^2}{2^2}

F=1.5625
$
\
Now we use R to find the critical value for the f test
\
The critical value for the f-test is between 0.362140511593323 and 2.76
\
Since 1.5625 is within the critical value we fail to reject the null hypothesis and we conclude that there is evidence for this two standard deviations to from the same population
\

\

$

\\
