## 2. In probability the concepts of independent events and mutually exclusive events are often confused. Clearly define each type of event, explain the difference between them, and give a good example for each.

**Independent events** are events where the probability of one of the events does not affect the probability that the other event occurs. In other words, the events do not influence the probability of each other occuring.
\
**For example**
The probability of obtaining tails when tossing a coin versus the probability of obtaininga 4 in a dice. In this case, the probability of obtaining heads does not affect the probability of obtaining a 4 in the dice therefore they are independent events.
\
P(A ∩ B )=P(A)*P(B)

**Mutually Exclusive Events**  are those events when the probability of one event occuring causes the probability of another event to not occur. In other words, if one event occurs then the other event cannot occur
\
**For example**
The probility of obtaining tails when tossing a coin versus the probability of obtaing heads is mutually exclusive since the probability of one occuring excludes the probability of the other occuring therefore the probability of both occuring is zero.
\
P(A ∩ B )=0

**Difference**
The main difference between the two is that mutually exlcusive  events are not independent and that independent events cannot be mutually exclusive



## 3. What is relative risk and how do you interpret it?

**Relative Risk** is the ratio of the probability of event B occuring given the probability of event A versus the probability of event B occuring given the probability of A not occurring.
In other words, the relative risk P(B|A)/P(B | $\hat{A}$ ). For example, the relative risk of developing lung cancer in non-smokers and smokers.
