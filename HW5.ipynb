## 2. A dermatologist investigating a certain type of skin cancer induced the disease in $n=16$ rats and then treated them with a new experimental drug. For each rat, the number of hours until remission of the cancer was recorded. The rats had a mean remission time of 400 hours and a standard deviation of 32 hours. 
### Calculate 95% confidence intervals for $\mu$ and $\sigma$.


Given that we do not know the population standard deviation we have to use a t distribution
\
T-distributions are used to estimate population parameters when the sample size is small and when the population variance is unknown.
\
\
**Confidence Interval for the Population mean**
\
n=16
\
Sample Mean=400
\
Sample Standard Deviation=32
\
$$\begin{equation}
\bar{X} \pm t_{0.025,15} \frac{s}{\sqrt{n}}
\\
400 \pm (2.131) \frac{32}{\sqrt{16}}
\\
\mu =400 \pm 17.048
\end{equation}$$
\
**Confidence Interval for the Standard deviation**
\
For this part we are going to use a chi-square distribution to approximate the population standard deviation
\
n=16
\
Sample Mean=400
\
Sample Standard Deviation=32
\
$$\begin{equation}
\sqrt{\frac{(n-1)s^2}{x^2_{\frac{\alpha}{2}}}} \leq \sigma \leq \sqrt{\frac{(n-1)s^2}{x^2_{1-\frac{\alpha}{2}}}}
\\
\sqrt{\frac{(16-1)(32)^2}{x^2_{\frac{\alpha}{2}}}} \leq \sigma \leq \sqrt{\frac{(16-1)(32)^2}{x^2_{1-\frac{\alpha}{2}}}}
\\
\sqrt{\frac{15360}{x^2_{\frac{\alpha}{2}}}} \leq \sigma \leq \sqrt{\frac{15360}{x^2_{1-\frac{\alpha}{2}}}}
\\
\sqrt{\frac{15360}{27.488}} \leq \sigma \leq \sqrt{\frac{15360}{7.261}}
\\
23.64 \leq \sigma \leq 45.99
\end{equation}$$

## 6. 50 adult black ducks from a large flock were trapped and weighed. Their mean weight was $\overline{x}=1500g$ and the variance of their weights was $s^2=8100g^2$. 
### Find the 99% confidence intervals for the mean, variance, and standard deviation of the flock.

```r
# Confidence Interval for the population mean
sm<-1500
sstd<-sqrt(8100)
n<-50
error<-qt(0.995, df= 49) * sstd/sqrt(n)
print('Confidence Interval for mean of the population')
sm + error 
sm - error

# Confidence Interval for the Standard Deviation mean
print('Confidence Interval for standard Deviation of the population')
sqrt(((n-1)*(sstd)^2)/qchisq(0.005,df=49, lower.tail=TRUE))
sqrt(((n-1)*(sstd)^2)/qchisq(0.995,df=49, lower.tail=TRUE))
# Confidence Interval for the Variance mean
print('Confidence Interval for variance of the population')
((n-1)*(sstd)^2)/qchisq(0.005,df=49, lower.tail=TRUE)
((n-1)*(sstd)^2)/qchisq(0.995,df=49, lower.tail=TRUE)
```

## 8. Out of 2500 cochlear implants manufactured in the month of September, 200 are tested for hermetic seals. The results showed that 35 implants failed the tests. 
### Calculate the 95% confidence limits for the proportion of defective implants in the batch (population). 


```r
p<- 35/200
q<- 1-p
n<- 200
z=1.96
p+z*sqrt((p*q)/n)
p-z*sqrt((p*q)/n)
```

## 9. Use R to plot a standard normal distribution from $x=-5$ to $x=5$ as well as the Student's $t$-distribution with $df = (5, 10, 15, 20, 25, 30)$. Use black for the normal distribution and different colors for each $t$-distribution.

```r
xaxis<-seq(-5,5,by=0.02)
yaxis<-dnorm(xaxis)
plot(xaxis,yaxis,type="l", xlab="x value",ylab="Density", main="Comparison of t Distributions")
degf <-c(seq(5,30,5))
color<-c("green","blue","yellow","pink","gold","orange","black")
labels<- c("df=5", "df=10", "df=15", "df=20","df=25","df=30" ,"normal")
lines(xaxis, dt(xaxis,5), lwd=2, col="green")
lines(xaxis, dt(xaxis,10), lwd=2, col="blue")
lines(xaxis, dt(xaxis,15), lwd=2, col="yellow")
lines(xaxis, dt(xaxis,20), lwd=2, col="pink")
lines(xaxis, dt(xaxis,25), lwd=2, col="gold")
lines(xaxis, dt(xaxis,30), lwd=2, col="orange")
legend("topright", inset=.05, title="Distributions",labels, lwd=2, lty=c(1, 1, 1, 1, 2), col=color)
```
